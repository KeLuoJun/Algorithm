```python
def sort():
    for i in range(n-1):
        for j in range(n-i-1):
            if t[j] > t[j + 1]:
                t[j], t[j + 1] = t[j + 1], t[j]
                c[j], c[j + 1] = c[j + 1], c[j]


def check(mid, n, m, t, c):
    for i in range(n - 1, -1, -1):
        if t[i] > mid:
            m -= c[i] * (t[i] - mid)
            t[i] = mid
            if m <= 0:
                return False
    return m >= 0

def solution(n, m, k, t, c):
    l, r = k, max(t)
    ans = -1
    while l <= r:
        mid = (l + r) // 2
        if check(mid, n, m, t, c):
            ans = mid
            r = mid - 1
        else:
            l = mid + 1
    return ans


if __name__ == '__main__':
    n, m, k = map(int, input().split())
    t = [0] * n
    c = [0] * n
    for i in range(n):
        t[i], c[i] = map(int, input().split())
    # sort()
    print(solution(n, m, k, t, c))
```

```python
def simplified_transformer():
    # import sys
    # input = sys.stdin.read
    # data = input().splitlines()

    # 读取矩阵大小
    n, d = map(int, input().split())

    # 读取矩阵 Q
    Q = []
    for i in range(1, n + 1):
        Q.append(list(map(int, input().split())))

    # 读取矩阵 K
    K = []
    for i in range(n + 1, 2 * n + 1):
        K.append(list(map(int, input().split())))

    # 读取矩阵 V
    V = []
    for i in range(2 * n + 1, 3 * n + 1):
        V.append(list(map(int, input().split())))

    # 读取向量 W
    W = list(map(int, input().split()))

    # 转置 K 矩阵
    K_T = [[K[j][i] for j in range(n)] for i in range(d)]

    # 矩阵相乘 Q × K^T
    QK_T = [[sum(Q[i][k] * K_T[j][k] for k in range(d)) for j in range(n)] for i in range(n)]

    # 点乘 W
    for i in range(n):
        for j in range(n):
            QK_T[i][j] *= W[i]

    # 矩阵相乘 (Q × K^T) × V
    result = [[sum(QK_T[i][k] * V[k][j] for k in range(n)) for j in range(d)] for i in range(n)]

    # 输出结果
    for row in result:
        print(" ".join(map(str, row)))

simplified_transformer()
```

```python
import numpy as np

class Relation:
    def __init__(self, size):
        self.size = size
        self.matrix = np.zeros((size, size), dtype=int)
    
    def add_relation(self, i, j):
        """Add relation (i,j) to matrix"""
        self.matrix[i][j] = 1
    
    def is_reflexive(self):
        """Check if relation is reflexive"""
        for i in range(self.size):
            if self.matrix[i][i] != 1:
                return False
        return True
    
    def is_symmetric(self):
        """Check if relation is symmetric"""
        for i in range(self.size):
            for j in range(self.size):
                if self.matrix[i][j] != self.matrix[j][i]:
                    return False
        return True
    
    def is_antisymmetric(self):
        """Check if relation is antisymmetric"""
        for i in range(self.size):
            for j in range(self.size):
                if i != j and self.matrix[i][j] == 1 and self.matrix[j][i] == 1:
                    return False
        return True
    
    def is_transitive(self):
        """Check if relation is transitive"""
        for i in range(self.size):
            for j in range(self.size):
                for k in range(self.size):
                    if self.matrix[i][j] == 1 and self.matrix[j][k] == 1:
                        if self.matrix[i][k] != 1:
                            return False
        return True
    
    def is_irreflexive(self):
        """Check if relation is irreflexive"""
        for i in range(self.size):
            if self.matrix[i][i] == 1:
                return False
        return True
    
    def warshall_closure(self):
        """Calculate transitive closure using Warshall algorithm"""
        closure = np.copy(self.matrix)
        for k in range(self.size):
            for i in range(self.size):
                for j in range(self.size):
                    closure[i][j] = closure[i][j] or (closure[i][k] and closure[k][j])
        return closure

def main():
    # Example usage
    r = Relation(4)
    # Add some relations
    relations = [(0,1), (1,2), (2,3), (1,1)]
    for i, j in relations:
        r.add_relation(i, j)
    
    print("Relation Matrix:")
    print(r.matrix)
    print("\nProperties:")
    print(f"Reflexive: {r.is_reflexive()}")
    print(f"Symmetric: {r.is_symmetric()}")
    print(f"Antisymmetric: {r.is_antisymmetric()}")
    print(f"Transitive: {r.is_transitive()}")
    print(f"Irreflexive: {r.is_irreflexive()}")
    
    print("\nTransitive Closure:")
    print(r.warshall_closure())

if __name__ == "__main__":
    main()
```
